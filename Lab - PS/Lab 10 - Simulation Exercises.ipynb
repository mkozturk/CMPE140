## Estimating pi

$$
\begin{aligned}
 \frac{\text{Points within circle}}{\text{All points}} &=& \frac{\pi*R^2}{2R*2R} \\
 \frac{\text{Points within circle}}{\text{All points}} &=& \frac{\pi}{4} \\
 \frac{4 * \text{Points within circle}}{\text{All points}} &=& \pi
\end{aligned}
$$

```r
estimate_pi <- function(x, y) {
    return(4*sum(x^2 + y^2 <= 1)/length(x))
}
```

## 100 Experiment with 100 Sample Points

```r
pi_est_100 <- c()
for (i in 1:100){
    x <- runif(100,-1,1)
    y <- runif(100,-1,1)
    pi_est_100 <- c(pi_est_100, estimate_pi(x, y))
}
```

```r
mean(pi_est_100)
median(pi_est_100)
sd(pi_est_100)
```

```r
hist(pi_est_100, breaks=20, col = "orange",
     main = "pi Estimations for 100 Experiment", xlab = "Estimations")
```

## Effect of Sample Size

We are going to try to estimate pi with different sample sizes. We are going to use sample sizes starting from 100 and increment by 100 until 10000. For each sample size, we are going to experiment 1000 times and change in standard deviation

```r
nsample_vec <- seq(100,10000,100)
nexperiments <- 1000
sdvals <- c()
for (nsample in nsample_vec){
    pi_est <- c()
    for (i in 1:nexperiments){
        x <- runif(nsample,-1,1)
        y <- runif(nsample,-1,1)
        pi_est <- c(pi_est, estimate_pi(x, y))
    }
    sdvals <- c(sdvals, sd(pi_est))
}
```

```r
# standard deviation value and its mirror
plot(nsample_vec, sdvals, type="l", ylim = c(-0.2, 0.2),
    xlab = "Sample Size", ylab = "Standard Deviation")
lines(nsample_vec, -sdvals, type="l")
# actual prediction line
lines(c(-250, 11000), c(0,0), col="red", lwd=2)
```
